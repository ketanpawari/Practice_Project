```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
Advertising_Data = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/DSData/master/Advertising.csv')
Advertising_Data.head()
```

```python
Advertising_Data.shape
```

```python
x = Advertising_Data[['TV','radio','newspaper']]
y = Advertising_Data['sales']
```

```python
# Splitting the data into training data and test data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=100)
```

```python
# Model training
# Linear Regression
mlr=LinearRegression()
```

```python
# training the Linear Regression model with training data
mlr.fit(x_train, y_train)
```

```python
print('Intercept :',mlr.intercept_)
print('Coefficient :')
list(zip(x,mlr.coef_))
```

```python
y_pred_mlr=mlr.predict(x_test)
print('Prediction for test Set : {}'.format(y_pred_mlr))
```

```python
mlr.df = pd.DataFrame({'Actual Value': y_test, 'Predicted Value': y_pred_mlr})
mlr.df
```

```python
mlr.predict([[56,55,67]])
```

```python
from sklearn import metrics
meanAErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))
print('R squared : {:0.2f}'.format(mlr.score(x,y)*100))
print('Mean Absolute Error:',meanAErr)
print('Mean Square Error:',meanSqErr)
print('Root Mean Absulute Error:',rootMeanSqErr)
```
